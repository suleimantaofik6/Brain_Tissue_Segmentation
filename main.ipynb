```python
import random
import matplotlib.pyplot as plt

def generate_random_numbers(num_points=10):
    return [random.random() for _ in range(num_points)]

def plot_random_numbers(numbers):
    plt.plot(numbers, marker='o', linestyle='-', color='b')
    plt.title('Random Numbers Plot')
    plt.xlabel('Index')
    plt.ylabel('Random Number')
    plt.show()
```

```python
num_points = 10  # You can change this to the desired number of random points
random_numbers = generate_random_numbers(num_points)
plot_random_numbers(random_numbers)
```
